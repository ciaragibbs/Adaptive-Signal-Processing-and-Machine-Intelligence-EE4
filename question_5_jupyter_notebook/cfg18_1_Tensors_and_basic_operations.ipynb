# Tensors and basic operations

<img src="./imgs/tensors_dogs.jpg" alt="Drawing" style="width: 500px;"/>

A tensor is a multi-dimenaional array of data where each dimension is conventionally referred to as the **mode**. Tensor order is defined by the number of its modes which is equivivalent to the number of indices required to identify a particular entry of a multi-dimensional array. For ease of visualisation and compact notation, we consider a third order tensor $\mathbf{\underline{X}} \in \mathbb{R}^{I \times J \times K}$. Thus, an element of such a tensor, $\mathbf{\underline{X}} \in \mathbb{R}^{I \times J \times K}$, can be written in a general form as:

$$ x_{ijk} = \mathbf{\underline{X}}[i, j, k]$$

> **Note:** In order to be consistent with Python indexing, count of modes and elements within starts from zeros.

Creation of a tensor starts with a formation of a multidimensional array of data.

```python
# Create 3-d array of data
array_3d = np.arange(24).reshape((2, 3, 4))

# Create tensor
tensor = Tensor(array_3d)

# Result preview
print(tensor)
tensor.data
```

### Solution: Part 4

Firstly, from Part 2 of Assignment 1, it is clear that as we vary the order of a tensor ($O$), but assign a fixed equal value to all dimensions that compose this order ($D$), the relationship between order and dimension in this particular case is $D^{O}$. With the numbers provided this would be $2^{O}$, where $O$ ranged from 3 to 20.
From Part 3 of Assignment 1, when instead the order of the tensor remains fixed, but the dimensions that compose this order are varied, the general relationship that underpins this case is again $D^{O}$, but now the power remains fixed. Hence, with the numbers provided this would be $D^{3}$, where $D$ range from 3 to 15.

Therefore, we see that an increase in either 1) order with a fixed dimension or 2) dimension with a fixed order results in an exponential and cubic increase in the number of elements actually required to construct the desired tensor,respectively. When analysing the curves more closely and plotting them in overlay as seen below, we can see that when the iterator i is greater or equal to 10, $2^{i} > i^{3}$, hence our relationship when order varies with a fixed dimension becomes greater than when dimension varies with a fixed order. When the order is greater than 10, we see that the effect of a non-constant exponent results in a exponential, substantial rise in elements required, much greater than when the exponent was fixed by the order of 3 i.e. cubic relationship.Hence the case where order varies with a fixed dimension, suffers more under the 'Curse of Dimensionality', a term coined by Bellman to describe scenarios in which an increase in the degress of freedom of data results in rapid increases in the number of parameters required to describe that data. From a practical perspective, this means that the memory needed to store such a tensor would also increase exponentially and subsequently for high-order tensors with the fixed dimension, we may lack sufficient storage space, not just for holding the values, but even for later tensor manipulations.



```python
all_elements_2 = []
for i in range(3,21):
    
    # order now remains at 3 but dimensions change
    tensor = Tensor(np.ones(tuple(i*np.ones(3).astype('int'))))
    all_elements_2.append(i**3)
    # or alternatively all_elements.append(tensor.size)

plt.plot(np.arange(3,21),all_elements_1)
plt.plot(np.arange(3,21),all_elements_2)
plt.xlabel('Tensor Dimensions')
plt.ylabel('Number of elements')
plt.title('Tensor Dimensions vs Number of Elements for Tensor Order 3')
plt.show()
```

# Unfolding

Conventionally, unfolding is considered to be a process of element mapping from a tensor to a matrix. In other words, it arranges the mode-$n$ fibers of a tensor to be the matrix columns. Thus, the mode-$n$ unfolding is denoted as:

$$\mathbf{\underline{A}} \xrightarrow{n} \mathbf{A}_{(n)}$$

Thus, this operations requires to specify a mode along which a tensor will be unfolded. For a third order tensor, a visual representation of such operation is as following

<img src="./imgs/unfolding.png" alt="Drawing" style="width: 600px;"/>

In `hottbox` this functionality is available through the corresponding methods of the `Tensor` object, for instance:

```python
# Create 3-d array of data
array_3d = np.arange(24).reshape((2, 3, 4))

# Create tensor
tensor = Tensor(array_3d)

# Unfolding of a tensor
tensor.unfold(mode=0)

# Result preview
print(tensor)
tensor.data
```

# Mode-n product 
<img src="./imgs/mode_n_product.png" alt="Drawing" style="width: 600px;"/>

The mode-$n$ product is a multiplication of a tensor  by a matrix along the $n^{th}$ mode of a tensor. This essentially means that each mode-$n$ fiber should be multiplied by this matrix. Mathematically, this can be expressed as:

$$\mathbf{\underline{X}} \times_n \mathbf{A} = \mathbf{\underline{Y}} \quad \Leftrightarrow  \quad \mathbf{Y}_{(n)} = \mathbf{A} \mathbf{X}_{(n)}  $$

```python
I, J, K = 2, 3, 4
J_new = 5

# Create tensor
array_3d = np.arange(I * J * K).reshape(I, J ,K)
X = Tensor(array_3d)

# Create matrix
A = np.arange(J_new * J).reshape(J_new, J)

# Performs Mode-n product 
Y = X.mode_n_product(A, mode=1, inplace=False)

# Preview of resulting tensor
print(Y)

# Preview of its unfolded version
Y.unfold(mode=1, inplace=False).data
```

# Properties of mode-n product

1. For distinct modes in a series of multiplications, the order of the multiplication is irrelevent: 

    $$\mathbf{\underline{X}} \times_n \mathbf{A} \times_m \mathbf{B} = \mathbf{\underline{X}} \times_m \mathbf{B} \times_n \mathbf{A} \quad (m \neq n)$$

1. However, this does not hold if the modes are the same :

    $$\mathbf{\underline{X}} \times_n \mathbf{A} \times_n \mathbf{B} = \mathbf{\underline{X}} \times_n (\mathbf{B}\mathbf{A})$$

### Solution: Part 2

Use [LaTeX syntax](https://stackoverflow.com/questions/13208286/how-to-write-latex-in-ipython-notebook) for analytical deriviations for the second property below.



Consider an  $N$-order tensor $\mathbf{\underline{X}} \in \mathbb{R}^{I_{1}\times I_{2}\times \dots \times I_{N} }$, where $I_{N}$ indexes each mode. Therefore a given element of the tensor $\mathbf{\underline{X}} \in \mathbb{R}^{I_{1}\times I_{2}\times \dots \times I_{N} }$, can be written in a general (coded notation) form as:

$$ x_{ijk} = \mathbf{\underline{X}}[i_{1}, i_{2}, \dots,i_{n}]$$

The n-mode product is defined as the multiplication of a tensor by a matrix along the $n^{th}$ mode of the tensor. Hence, each mode-$n$ fiber is to be multiplied by this matrix:

$$\mathbf{\underline{X}} \times_n \mathbf{A} = \mathbf{\underline{Y}} \quad \Leftrightarrow  \quad \mathbf{Y}_{(n)} = \mathbf{A} \mathbf{X}_{(n)}  $$ 

where the right hand side represents the n-mode product in terms of unfolded tensors. The size of $\mathbf{Y}_{(n)}$ (the result of the n-mode product) given the dimensions of the tensor $\mathbf{\underline{X}} \in \mathbb{R}^{I_{1}\times I_{2}\times \dots \times I_{N} }$, and matrix $\mathbf{A} \in \mathbb{R}^{J_{n}\times I_{n}}$ is $I_{1} \times \dots \times I_{n-1} \times J \times I_{n+1} \times \dots \times I_{N}$. This is defined in an element-wise manner below:
 
\begin{equation}
    \mathbf{\underline{Y}}_{ (i_{1}, \dots , i_{n-1} , j_{n}  , i_{n+1} , \dots , i_{N})} = (\mathbf{\underline{X}} \times_n \mathbf{A})_{ (i_{1}, \dots , i_{n-1} , j_{n}  , i_{n+1} , \dots , i_{N})}
   = \sum_{i_{n}=1}^{I_{n}} x_{(i_{1}, i_{2}, \dots, i_{n}, \dots , i_{N})} a_{(j_{n},i_{n})}
\end{equation}

With the expanded element-wise notation above, we now prove the second provided statement, which applies for equal modes: $$\mathbf{\underline{X}} \times_n \mathbf{A} \times_n \mathbf{B} = \mathbf{\underline{X}} \times_n (\mathbf{B}\mathbf{A})$$ 


For the left hand side:

\begin{equation}
    \mathbf{\underline{Yleft}}
   = \sum_{j_{n}=1}^{J_{n}}\bigg(\sum_{i_{n}=1}^{I_{n}} x_{(i_{1}, i_{2}, \dots, i_{n}, \dots , i_{N})} a_{(j_{n},i_{n})}\bigg)b_{(k_{n},j_{n})} = \sum_{i_{n}=1}^{I_{n}} x_{(i_{1}, i_{2}, \dots, i_{n}, \dots , i_{N})} \sum_{j_{n}=1}^{J_{n}} \bigg( a_{(j_{n},i_{n})}b_{(k_{n},j_{n})} \bigg)
\end{equation}

the above can also be rearranged (due to the summation):
\begin{equation}
    \mathbf{\underline{Yleft}}
   = \sum_{i_{n}=1}^{I_{n}} x_{(i_{1}, i_{2}, \dots, i_{n}, \dots , i_{N})} \sum_{j_{n}=1}^{J_{n}} \bigg( b_{(k_{n},j_{n})}a_{(j_{n},i_{n})} \bigg)
\end{equation}
on the right hand side of this equation, the multiplication of elements $a_{(j_{n},i_{n})}$ and $b_{(k_{n},j_{n})}$ from matrices $\mathbf{A}$ and $\mathbf{B}$ respectively can be rewritten in element-wise notation of a new matrix $\mathbf{Z}$, where $\mathbf{Z} = \mathbf{B}\mathbf{A}$:

\begin{equation}
    \mathbf{\underline{Yleft}}
   = \sum_{i_{n}=1}^{I_{n}} x_{(i_{1}, i_{2}, \dots, i_{n}, \dots , i_{N})}z_{(k_{n},i_{n})} = \mathbf{\underline{Yleft}}_{ (i_{1}, \dots , i_{n-1} , k_{n}  , i_{n+1} , \dots , i_{N})}
\end{equation}

If we now convert from the element-wise notation to the matrix n-mode product notation, we see that our rearranged formulation of the left hand side gives:

\begin{equation}
\mathbf{\underline{X}} \times_n \mathbf{Z} = \mathbf{\underline{X}} \times_n \mathbf{BA} = \mathbf{\underline{Yright}}
\end{equation}

Hence since the left hand side has been successfully rearranged to match the formulation of the right hand side, we thereby prove the second statement provided holds true.
